```python
import pandas as pd
properties = pd.read_csv('properties.csv', index_col = 'id')
rooms = pd.read_csv('rooms.csv', index_col = 'id')
```

```python
propC = properties.status.isin(['available','partial'])
properties = properties[propC]
```

```python
propInd = pd.DataFrame(properties.index, columns= ['id'], index = properties.index)
properties = pd.concat([propInd, properties], axis = 1)
```

```python
localities_DF = pd.get_dummies(properties.locality)
city_DF = pd.get_dummies(properties.city)
```

```python
localities_DF
```

```python
import numpy as np
areas_DF = pd.DataFrame(np.zeros((properties.shape[0],3), dtype = int), columns = ['areaLessThan1000','areaLessThan1700', 'areaGreaterThan1700'], index = properties.index)
for i in properties.index:
    if(properties.at[i,'sqftsuper'] <= 1000):
        areas_DF.at[i, 'areaLessThan1000'] = 1
    elif(properties.at[i,'sqftsuper'] <= 1700):
        areas_DF.at[i, 'areaLessThan1700'] = 1
    else:
        areas_DF.at[i, 'areaGreaterThan1700'] = 1
areas_DF
```

```python
properties = properties.drop(['id','locality','city','owner_id', 'name', 'flat_no', 'addressl1', 'state',
       'country', 'zipcode','image', 'images', 'type','sharetype','availfrom', 'availto','details',
       'notes', 'cntrestaurants', 'cnthospitals', 'cntgyms', 'cnttheater',
       'cntmalls', 'workdistance', 'blockname','tds', 'actualOSAmount',
       'pendingOSAmount', 'amountRemark', 'minimumGuranteeIsActive',
       'minimumGuranteeAmount', 'minimumGuranteeStart', 'minimumGuranteeEnd',
       'propertyrules', 'parking_number', 'metatitle', 'metakeyword',
       'metadescription', 'assign_to','partner', 'showinhomepage',
       'created_at', 'updated_at', 'owner_id2', 'owner_id3','tenure',
       'latitude', 'longitude','gender', 'tag_as', 'commpercentage','furnished','owner_rent','status'], axis = 1) 
```

```python
properties = pd.concat([localities_DF, city_DF, properties], axis = 1)
```

```python
properties.columns
```

```python
x = properties.drop(columns = ['rent'], axis  = 1)
y = properties.rent
```

```python
x.columns
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(x_train, y_train)
```

```python
y_predict = reg.predict(x_test) 
```

```python
y_predict
```

```python
y_test
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_predict)  
```

```python
from sklearn.linear_model import Ridge
rid = Ridge(alpha=1000)
rid.fit(x_train, y_train)
y_ridge_pred = rid.predict(x_test)
```

```python
from sklearn.linear_model import Ridge
rid2 = Ridge(alpha=1000)
rid2.fit(x_train, y_train)
y_ridge_pred = rid2.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_ridge_pred) 
```
